## 2. Audit éthique complet (~1 h)

Calcul du **disparate impact** sur 3 variables sensibles + 1 intersection.

```python
def disparate_impact(df: pd.DataFrame, sensible: str, target: str = "income",
                     positive: str = ">50K") -> tuple[float, pd.Series]:
    """Calcule le DI = SR_min / SR_max + retourne la série des SR par groupe."""
    sr = df.groupby(sensible)[target].apply(lambda x: (x == positive).mean())
    di = sr.min() / sr.max()
    return di, sr
```

```python
# TODO — DI sur sex
# di_sex, sr_sex = disparate_impact(df, 'sex')

# Fonction générique de calcul du DI
def get_di_analysis(dataframe, group_col, protected_value, privileged_value):
    prob_protected = (dataframe[dataframe[group_col] == protected_value]['income'] == '>50K').mean()
    prob_privileged = (dataframe[dataframe[group_col] == privileged_value]['income'] == '>50K').mean()
    di_ratio = prob_protected / prob_privileged
    return prob_protected, prob_privileged, di_ratio

# Nettoyage strict des chaînes
df['income'] = df['income'].astype(str).str.strip()
df['sex'] = df['sex'].astype(str).str.strip()


# 1. Calcul pour le Sexe (Femme vs Homme)
p_female, p_male, di_sex = get_di_analysis(df, 'sex', 'Female', 'Male')



print("========================================================================")
print("                   RAPPORT DISPARATE IMPACT (INCOME)                    ")
print("========================================================================")
print(f"[SEXE] Taux Femme (>50K) : {p_female*100:.2f}%")
print(f"[SEXE] Taux Homme (>50K) : {p_male*100:.2f}%")
print(f"--> DI (Femme vs Homme)  : {di_sex:.4f}  | [ALERTE SEVERE < 0.80]")
```

```python
# TODO — DI sur race
di_race, sr_race = disparate_impact(df, 'race')

# Nettoyage et analyse détaillée par rapport au groupe de référence 'White'
df['race'] = df['race'].astype(str).str.strip()
p_white = (df[df['race'] == 'White']['income'] == '>50K').mean()

print("========================================================================")
print("                   RAPPORT DISPARATE IMPACT (RACE)                      ")
print("========================================================================")
print(f"[RACE] Taux Référence 'White' (>50K) : {p_white*100:.2f}%")

for r in df['race'].unique():
    if r != 'White':
        # Utilisation de votre fonction unitaire pour le détail
        p_r, _, di_r = get_di_analysis(df, 'race', r, 'White')
        status = "[CONFORME >= 0.80]" if di_r >= 0.80 else "[ALERTE SEVERE < 0.80]"
        print(f"--> Groupe '{r:<19}' : Taux = {p_r*100:.2f}% | DI vs White = {di_r:.4f} {status}")
print("========================================================================")
```

```python
# TODO — DI sur native_country (agrégé en USA / non-USA pour simplifier)
# df['native_us'] = (df['native_country'] == 'United-States').map({True: 'USA', False: 'non-USA'})


# Agrégation de la variable native_country en USA / non-USA
df['native_us'] = (df['native_country'] == 'United-States').map({True: 'USA', False: 'non-USA'})

# TODO — DI sur native_country
di_country, sr_country = disparate_impact(df, 'native_us')

# Affichage des résultats
print(f"Disparate Impact (Origine) : {di_country:.4f}")
print("\nTaux de sélection par origine géographique :")
print(sr_country)
```

### Intersectionnalité — DI sur le croisement sex × race

```python
# TODO — créer une colonne sex_race et calculer le DI
# df['sex_race'] = df['sex'].str.cat(df['race'], sep='_')

# Nettoyage préventif et création de la colonne intersectionnelle sex_race
df['sex_race'] = df['sex'].str.strip().str.cat(df['race'].str.strip(), sep='_')

# TODO — Calcul du DI sur la variable croisée
di_inter, sr_inter = disparate_impact(df, 'sex_race')

print("========================================================================")
print(f"   RÉSULTAT DE L'AUDIT INTERSECTIONNEL (Sex × Race) | DI : {di_inter:.4f}   ")
print("========================================================================")
print("Taux de sélection par combinaison (ordonné du plus bas au plus haut) :")
print(sr_inter.sort_values().to_frame(name='Taux >50K').to_string())
print("========================================================================")
```

## 3. Visualisations (≥ 5)

Au moins 5 visualisations : distribution cible + DI par variable sensible +
crosstab intersection.

```python
# TODO — 5 visualisations claires

import matplotlib.pyplot as plt
import seaborn as sns

# Configuration du style global pour le rapport
sns.set_theme(style="whitegrid")
fig = plt.figure(figsize=(20, 14))
plt.suptitle("TÂCHE 2 — AUDIT GRAPHIQUE ÉTHIQUE COMPLET", fontsize=18, fontweight='bold', y=0.98)

# ========================================================================
# GRAPH-1 : Distribution de la Cible (Income)
# ========================================================================
ax1 = plt.subplot2grid((3, 2), (0, 0))
target_pct = df['income'].value_counts(normalize=True) * 100
sns.barplot(x=target_pct.index, y=target_pct.values, palette='Blues_r', ax=ax1)
ax1.set_title("1. Déséquilibre de la Variable Income", fontsize=12, fontweight='bold')
ax1.set_ylabel("Pourcentage (%)")
ax1.set_ylim(0, 100)
for i, v in enumerate(target_pct.values):
    ax1.text(i, v + 2, f"{v:.1f}%", ha='center', fontweight='bold')

# ========================================================================
# GRAPH-2 : Disparate Impact — Sexe
# ========================================================================
ax2 = plt.subplot2grid((3, 2), (0, 1))
sns.countplot(data=df, x='sex', hue='income', palette='Set2', ax=ax2)
ax2.set_title(f"2. Focus Sexe — Ratio DI calculé : {di_sex:.3f}", fontsize=12, fontweight='bold')
ax2.set_xlabel("Sexe")
ax2.set_ylabel("Effectif")
# Annotation du statut légal
status_sex = "ALERTE < 0.80" if di_sex < 0.8 else "CONFORME"
ax2.text(0.5, df['sex'].value_counts().max()*0.8, status_sex, color='red' if di_sex < 0.8 else 'green', 
         fontsize=14, fontweight='bold', bbox=dict(facecolor='white', alpha=0.8), ha='center')

# ========================================================================
# GRAPH-3 : Disparate Impact — Race
# ========================================================================
ax3 = plt.subplot2grid((3, 2), (1, 0))
sns.countplot(data=df, x='race', hue='income', palette='Set2', ax=ax3)
ax3.set_title(f"3. Focus Race — Ratio DI global min/max : {di_race:.3f}", fontsize=12, fontweight='bold')
ax3.tick_params(axis='x', rotation=15)
ax3.set_xlabel("Origine Ethnique")
ax3.set_ylabel("Effectif")

# ========================================================================
# GRAPH-4 : Disparate Impact — Nationalité (native_us)
# ========================================================================
ax4 = plt.subplot2grid((3, 2), (1, 1))
sns.countplot(data=df, x='native_us', hue='income', palette='Set2', ax=ax4)
ax4.set_title(f"4. Focus Pays d'Origine — Ratio DI calculé : {di_country:.3f}", fontsize=12, fontweight='bold')
ax4.set_xlabel("Statut Géographique")
ax4.set_ylabel("Effectif")
status_country = "ALERTE < 0.80" if di_country < 0.8 else "CONFORME"
ax4.text(0.5, df['native_us'].value_counts().max()*0.8, status_country, color='red' if di_country < 0.8 else 'green', 
         fontsize=14, fontweight='bold', bbox=dict(facecolor='white', alpha=0.8), ha='center')

# ========================================================================
# GRAPH-5 : Croisement Intersectionnel (Crosstab Sexe × Race)
# ========================================================================
ax5 = plt.subplot2grid((3, 2), (2, 0), colspan=2)

# Construction d'un tableau croisé d'occurrences pour la heatmap
ct = pd.crosstab(df['sex'], df['race'], values=df['income'].apply(lambda x: 1 if x == '>50K' else 0), aggfunc='mean') * 100

sns.heatmap(ct, annot=True, fmt=".1f", cmap="YlOrRd", cbar_kws={'label': 'Taux de hauts revenus (%)'}, ax=ax5, annot_kws={"size": 12, "weight": "bold"})
ax5.set_title(f"5. Crosstab Intersectionnel (Sex × Race) — Taux d'accès au revenu >50K (DI global : {di_inter:.3f})", fontsize=13, fontweight='bold')
ax5.set_xlabel("Race")
ax5.set_ylabel("Sexe")

plt.tight_layout()
plt.show()
```
